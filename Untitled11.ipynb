# Lenguaje Python, crear la clase "animal" con los atributos "nombre", "raza" crear el constructor animal, y las funciones ladrar(), correr(),  crear los objetos pitbull, chihuahua, nivel principiante, no es interactivo, proporciona tu los datos

```python
# Definimos la clase (el molde)
class Animal:
    # El constructor: se ejecuta al crear el objeto y asigna los datos
    def __init__(self, nombre, raza):
        self.nombre = nombre  # Atributo nombre
        self.raza = raza      # Atributo raza

    # Función para que el animal ladre
    def ladrar(self):
        print(f"{self.nombre} ({self.raza}) dice: ¡Guau, guau!  🐱‍👤")

    # Función para que el animal corra
    def correr(self):
        print(f"{self.nombre} está corriendo velozmente.   🐱‍🏍")

    def morder(self):
        print(f"{self.nombre} está mordiendo a la vecina   🤱")

# --- Creación de los objetos (Instanciación) ---

# Creamos el objeto pitbull con datos específicos
pitbull = Animal("Rambo", "Pitbull")

# Creamos el objeto chihuahua con datos específicos
chihuahua = Animal("Chispita", "Chihuahua")
# creamdo el objeto perro callegero con datos
perrocallejero = Animal("solovino", "chafa")

# --- Pruebas de funcionamiento ---

print("--- Interacción con el Pitbull ---")
pitbull.ladrar()
pitbull.correr()

print("\n--- Interacción con el Chihuahua ---")
chihuahua.ladrar()
chihuahua.correr()

print("\n--- Interacción con el Perro Callejero ---")
perrocallejero.ladrar()
perrocallejero.correr()
perrocallejero.morder()
print("programa creado por David Barrera NL = 8")
```
